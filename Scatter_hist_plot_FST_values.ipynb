# This is a script I wrote in R to visualize the "FST" values for different locations of single nucleotide polymorphisms, and their statistical significance

```r
setwd("~/GBS/GBS3")
fst <- read.table("Fst_table_all.txt", sep="\t", header=T)

# SET FUNCTION
##### seperate them out
##Significant SNPs

## par= plot area
## oma= outer margin area
## cex= font size

fst <- read.table("Fst_table_all.txt", sep="\t", header=T)

#function
scatterhist1 = function(x, y, cutoff1=cutoff1, cutoff2=cutoff2, xlab="Fst", ylab="-LOG(p-value)", col='black', pch=pch, cex=cex, main=''){
  par(oma = c(4,4,2,1))
  zones=matrix(c(1,0,2,3), ncol=2, byrow=TRUE)
  layout(zones, widths=c(5/6, 1/6), heights=c(1/6, 5/6))
  xhist = hist(x, plot=FALSE, breaks=40)
  yhist = hist(y, plot=FALSE, breaks=60)
  top = max(c(xhist$counts, yhist$counts))
  
  # top histogram
  par(mar=c(0,0,0,0))
  barplot(xhist$counts, axes=FALSE, space=0)
  
  # scatter plot
  par(mar=c(0,0,0,0))
  plot(x,y, pch=pch, cex=cex, col=ifelse(y>cutoff1,"dodgerblue3", "lightslategrey"))
  abline(cutoff1,0, lty=1, col="dodgerblue3")
  mtext(xlab, side=1, line=2)
  mtext(ylab, side=2, line=2)
  
  
  
  # side histogram
  par(mar=c(0,0,0,0))
  barplot(yhist$counts, axes=FALSE, space=0, horiz=TRUE)
  
  title(main, outer = TRUE)
}

##run function
scatterhist1(fst$FST,fst$n_log_pvalue, cex=1, cutoff1=2.321, pch=1)


## 10% FDR
fst <- read.table("Fst_table_all.txt", sep="\t", header=T)

#function
scatterhist2 = function(x, y, cutoff1=cutoff1, cutoff2=cutoff2, xlab="Fst", ylab="-LOG(p-value)", col='black', pch=pch, cex=cex, main=''){
  par(oma = c(4,4,2,1))
  zones=matrix(c(1,0,2,3), ncol=2, byrow=TRUE)
  layout(zones, widths=c(5/6, 1/6), heights=c(1/6, 5/6))
  xhist = hist(x, plot=FALSE, breaks=40)
  yhist = hist(y, plot=FALSE, breaks=60)
  top = max(c(xhist$counts, yhist$counts))
  
  # top histogram
  par(mar=c(0,0,0,0))
  barplot(xhist$counts, axes=FALSE, space=0)
  
  # scatter plot
  par(mar=c(0,0,0,0))
  plot(x,y, pch=pch, cex=cex, col=ifelse(y>cutoff2,"seagreen3", "lightslategrey"))
  abline(cutoff1,0, lty=1, col="dodgerblue3")
  abline(cutoff2,0, lty=1, col="seagreen3")
  mtext(xlab, side=1, line=2)
  mtext(ylab, side=2, line=2)
  
  
  
  # side histogram
  par(mar=c(0,0,0,0))
  barplot(yhist$counts, axes=FALSE, space=0, horiz=TRUE)
  
  title(main, outer = TRUE)
}

scatterhist2(fst$FST,fst$n_log_pvalue, cex=1, cutoff1=2.321, cutoff2=3.896, pch=1)


## Bonferroni's Correction

fst <- read.table("Fst_table_all.txt", sep="\t", header=T)

#function
scatterhist3 = function(x, y, cutoff1=cutoff1, cutoff2=cutoff2, cutoff3=cutoff3, xlab="Fst", ylab="-LOG(p-value)", col='black', pch=pch, cex=cex, main=''){
  par(oma = c(4,4,2,1))
  zones=matrix(c(1,0,2,3), ncol=2, byrow=TRUE)
  layout(zones, widths=c(5/6, 1/6), heights=c(1/6, 5/6))
  xhist = hist(x, plot=FALSE, breaks=40)
  yhist = hist(y, plot=FALSE, breaks=60)
  top = max(c(xhist$counts, yhist$counts))
  
  # top histogram
  par(mar=c(0,0,0,0))
  barplot(xhist$counts, axes=FALSE, space=0)
  
  # scatter plot
  par(mar=c(0,0,0,0))
  plot(x,y, pch=pch, cex=cex, col=ifelse(y>cutoff3,"mediumvioletred", "lightslategrey"))
  abline(cutoff1,0, lty=1, col="dodgerblue3")
  abline(cutoff2,0, lty=1, col="seagreen3")
  abline(cutoff3,0, lty=1,col="mediumvioletred")
  mtext(xlab, side=1, line=2)
  mtext(ylab, side=2, line=2)
  
  
  
  # side histogram
  par(mar=c(0,0,0,0))
  barplot(yhist$counts, axes=FALSE, space=0, horiz=TRUE)
  
  title(main, outer = TRUE)
}

## run function
scatterhist3(fst$FST,fst$n_log_pvalue, cex=1, cutoff1=2.321, cutoff2=3.896, cutoff3=6.537, pch=1)




```
